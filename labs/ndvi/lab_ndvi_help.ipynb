## Some code to help with the labs

If we have installed correctly 'rasterio', the file should be read without error

```python
import rasterio

ds = rasterio.open('myfile.tif')
bands = ds.read()
```

Here we can query the metadata of the file, eg

```python
print(ds.crs)
```

We can check how many bands and the shape (row*col) that our image has.
Here we can see: 4 bands, and the image is 378*545 pixels

```python
print(bands.shape)
```

We can also check for instance the type:

```python
print(bands.dtype)
```

You can visualise directly in jupyter the image

```python
from matplotlib import pyplot

pyplot.imshow(bands[0], cmap='pink')
pyplot.show()
```
